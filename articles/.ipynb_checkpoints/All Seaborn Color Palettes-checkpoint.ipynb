```python
cmap_list = []
for cmap in color_palettes:
    if '_r' not in cmap:
        cmap_list.append(cmap)
cmap_list = sorted(cmap_list, key=lambda x : x.lower())
```

```python
def plot_all_cmap():
    row = len(cmap_list)
    ind = [list(range(12))]
    fig, axes = plt.subplots(row, 1, figsize=(15, 1*row))
    for cmap, ax in zip(cmap_list, axes.reshape(-1)):
        sns.heatmap(ind, linewidth=1, cmap=cmap, square=True, cbar=False, xticklabels=False, yticklabels=False, ax=ax)
        ax.axhline(y=0, color='k',linewidth=3)
        ax.axhline(y=1, color='k',linewidth=3)
        ax.axvline(x=0, color='k',linewidth=2)
        ax.axvline(x=12, color='k',linewidth=2)
        if cmap == 'twilight_shifted':
            cmap = 'twilight\n_shifted'
        elif cmap == 'nipy_spectral':
            cmap = 'nipy\n_spectral'
        ax.set_ylabel(cmap, rotation='horizontal', labelpad=36, fontsize=14)
        ax_r = ax.twinx()
        ax_r.set_ylabel(cmap + '_r', rotation='horizontal', labelpad=41, fontsize=14)
        ax_r.set_yticks([])
    fig.tight_layout()
    plt.subplots_adjust(wspace=0, hspace=0)
    plt.show()
```
